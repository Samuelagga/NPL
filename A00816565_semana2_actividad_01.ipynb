*   **Pregunta 1.**

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario.

```python
#Descargar la biblioteca para expresiones regulares, usar sub
import re
P1 = [re.sub(r'\n',r'', line) for line in docs]
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!".

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen.

Indica cuántos resultados obtuviste.



```python
# Buscar palabras que terminan con dos o más signos de exclamación
exclamaciones = re.findall(r'(\w+)(!{2,})', all_text)

# Imprimir palabras encontradas junto con la cantidad de signos de exclamación
for palabra in exclamaciones:
    cantidad_exclamacion = palabra[1].count('!')
    print(f"Palabra: {palabra[0]+palabra[1]}, Cantidad de signos de exclamación: {cantidad_exclamacion}")

# Imprimir el total de resultados
print("Total de resultados:", len(exclamaciones))
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
# Buscar palabras en mayusculas
mayusculas = re.findall(r'\b[A-Z]+\b', all_text)
```

```python
#Funcion para imprimir palabras en un arreglo y mostrar la longitud
def print_words(lista):
    for palabra in lista:
        print(palabra)
    print("Total de palabras encontradas:", len(lista))
```

```python
print_words(mayusculas)
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes.
   




```python
#Eliminar caracteres no alfabeticos y volver minusculas
limpieza = [re.sub(r'(?!\s)\W', r'', comment.lower()) for comment in P1]
#Eliminar espacions adicionales
limpieza = [re.sub(r'\s+',r' ', comment) for comment in limpieza]
```

```python
#Impresion de los primeros 10 comentarios
limpieza[0:10]
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus.

Es decir, al final de este proceso de tokenización, debes tener como resultado una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus.

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
# Tokenizar el corpus por palabras
corpus_tokenizado = [word_tokenize(oracion) for oracion in limpieza]

# Imprimir el corpus tokenizado y el total de tokens
for i, oracion_tokenizada in enumerate(corpus_tokenizado, start=1):
    num_tokens = len(oracion_tokenizada)
    print(f"Oración {i} ({num_tokens} tokens): {oracion_tokenizada}")

# Calcular el total de tokens en todo el corpus
total_tokens = sum(len(oracion_tokenizada) for oracion_tokenizada in corpus_tokenizado)
print("Total de tokens en todo el corpus:", total_tokens)
```

```python
#limpieza del corpus
clean_corpus = []
for comment in corpus_tokenizado:
    clean_comment = [word for word in comment if word not in mis_stopwords]
    clean_corpus.append(clean_comment)
```

```python
#Calcular tokens restantes despues de limpieza
tokens_limpios = sum(len(comment) for comment in clean_corpus)
print("Total de tokens en el corpus limpio: ", tokens_limpios)
```

```python
#Calcular tokens individuales
vocab = set()
for comment in clean_corpus:
    vocab.update(comment)
tokens_unicos = len(vocab)
print("Total de tokens unicos en el corpus limpio:", tokens_unicos)
```
